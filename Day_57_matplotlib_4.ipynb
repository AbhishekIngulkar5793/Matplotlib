## pieplot: 
- univariate plot
- used to display % wise distribution of data
- used for categorywise comparison

```python
team = ['Ganesh','Seema','Abhishek']
captial = [12,10,20] #Cr.
```

```python
plt.pie(captial)
plt.show()
```

```python
team
```

```python
# display name of each team mate on pie
plt.pie(captial,labels=team)
plt.show()
```

```python
# add percentage on pie occupied by each member from capital
# autopct = '%%f%%'
plt.pie(captial,labels=team,autopct='%.f%%')
plt.show()
```

```python
# highlight /separate pie for Seema
plt.pie(captial,labels=team,autopct='%.2f%%',explode=[0.1,0.2,0.1],
       pctdistance=1.3,labeldistance=.4)
plt.show()
```

## subplot in python
- when we want to combine multiple plots in a single canvas
- so we can have lets say 2 plots with difffent axis and its existance then use subplot option

```python
display(team,captial)
```

```python
plt.subplot(121)#1(row) 2(columns) 1(plot)
plt.bar(team,captial)
plt.subplot(122)
plt.plot(team,captial)
```

```python
plt.subplot(211)#2(rows) 1(column) 1(plot)
plt.bar(team,captial)
plt.subplot(212)
plt.plot(team,captial)
```

```python
# 2 x 2 subplot
plt.subplot(221)
plt.subplot(222)
plt.subplot(223)
plt.subplot(224)
```

```python
plt.subplot(221)
plt.bar(team,captial)

plt.subplot(222)
plt.pie(captial,autopct='%.f%%',labels=team)

plt.subplot(223)
plt.scatter(team,captial)

plt.subplot(224)
plt.plot(a1,color='r')
```

```python
#check impact of tight_layout
plt.subplot(221)
plt.subplot(222)
plt.subplot(223)
plt.subplot(224)
plt.tight_layout(pad=2.08)
```

```python
# SET LABELS TO EACH PLOT
#plt.figure(figsize=(18,16))
plt.subplot(221)
plt.bar(team,captial)
plt.xlabel('Data')
plt.ylabel('Capitla per person')
plt.title('Bar plot')


plt.subplot(222)
plt.pie(captial,autopct='%.f%%',labels=team)
plt.xlabel('Data')
plt.ylabel('Capitla per person')
plt.title('Pie plot')


plt.subplot(223)
plt.scatter(team,captial)
plt.xlabel('Data')
plt.ylabel('Capitla per person')
plt.title('Scatter plot')

plt.subplot(224)
plt.plot(a1,color='r')
plt.xlabel('Data')
plt.ylabel('Capitla per person')
plt.title('Line plot')
plt.suptitle('Combining 4 plots[2x2]')
plt.tight_layout()
```
