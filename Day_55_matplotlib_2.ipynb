```python
import matplotlib.pyplot as plt
```

```python
values = [12,34,22,33,45,65,45,21,21,22,34,32,31]
```

```python
# univariate: only single input is used 
# the index is on X-axis
# and the values are on Y-axis
plt.plot(values,marker='X',markersize=8)
```

```python
plt.plot(values,marker='X',markersize=8,color='g',linestyle=':')
plt.xlabel('Data')
plt.ylabel('Count')
plt.title('Simple Graph')
```

```python
# int Q. How to save image of matplotlib graphs?
plt.plot(values,'*m-.',markersize=8,linewidth=1)
plt.xlabel('Data')
plt.ylabel('Count')
plt.legend('Simple Graph')
plt.savefig('Sample_graph.png',dpi=300) # dpi= dots per inch 
```

## Scatter plot
- It is used to display distribution of data in terms of circles/markers
- It is completely bivariate plot
- It is used to check outliers in data
- Outliers are an exceptional cases among data present
- a data point far away from the distribution is an outlier

```python
import numpy as np
```

```python
# P.S.- plot overs vs runs 
# take 10 overs data 
overs = range(1,11)
runs = [0,2,10,4,16,20,1,4,6,8]
```

```python
# we want runs in cumulative order
runs = np.cumsum(runs)
runs
```

```python
#plt.scatter(overs,runs,color='k')
plt.plot(overs,runs,color='k') # line
plt.plot(overs,runs,color='cyan',marker='x',linestyle=' ',markersize=12) # markers
```

```python
# fall of wickets
f_w = [None,  2, None, None, 32, None, None, None, None, 71]
```

```python
plt.plot(overs,runs,label='runs per over')
plt.scatter(overs,f_w,color='red',label='wickets')
plt.xlim((1,9))
plt.ylim([0,71])
plt.legend()
```
