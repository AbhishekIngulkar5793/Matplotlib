### line plot

```python
# lets see a line plot of Day vs Temperature
day = [1,2,3,4,5,6,7]
temp = [15,16,18,10,19,20,6]
```

```python
import matplotlib.pyplot as plt
```

```python
# line plot: we can use to show the progress of data, to show the performance,
# continuous changes in data, to check flow of something, 
# plt.plot(input,output)
plt.plot(day,temp)
```

```python
# 1. lets change the color
plt.plot(day,temp,color='black')
# choice of colors: white,black,magenta,yellow,cyan,green,red
```

```python
# 2. now lets use some colors with their hex value from https://html-color.codes/
plt.plot(day,temp,color='#bbe5ab')
```

```python
# 3. lets add a new linestyle
plt.plot(day,temp,color='#bbe5ab',linestyle='dashdot',linewidth= 4)
#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# now add labels to the plot
plt.plot(day,temp,color='#bbe5ab',linestyle=':',linewidth=4)
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('weekly data of Temperature')
plt.grid()
```

```python
# Multivariate plot
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 300

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a',fontsize=30)
plt.ylabel('entry b')
plt.title('sample from interenet')
plt.show()
```

```python
# to mark intersection point
# use marker attribute
# add labels to plot
plt.plot(day,temp,color='g',linestyle='',linewidth=9,marker='*')
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('Weekly data of Temp')
plt.grid()
```

```python
# now lets add a plot label
plt.plot(day,temp,color='red',linestyle='',linewidth=6,marker='o',label='Daywise Temperature')
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.title('Weekly data of Temp')
plt.grid()
plt.legend(loc=8)
```
